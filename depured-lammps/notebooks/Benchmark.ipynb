```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
times = np.array([234.5, 216 ,131.25 ,97.5 ,67.25 ,56.25 ,53, 60.5, 130.75])
ns = [1,2,4,6,8,10,12,16,20]
```

```python
plt.figure(num=None, figsize=(8,6))
plt.xlabel("Cores used")
plt.ylabel("Time spent (s)")
plt.title("Average time spent for 4 separate LAMMPS runs of 100k fs (10fs timestep)")
plt.xticks(ns)
plt.plot(ns,times, '.-');
```

```python
plt.figure(num=None, figsize=(8,6))
plt.xlabel("Cores used")
plt.ylabel("T1/Tn")
plt.title("Average time spent for 4 separate LAMMPS runs of 100k fs (10fs timestep)")
plt.xticks(ns)
plt.plot(ns,times[0]/times, '.-', label="LAMMPS");
plt.plot(ns,ns, '-', label="Ideal");
plt.legend();
```
