```python
int_pot = pd.DataFrame(np.zeros(shape=(len(dicti), len(dicti))), index=dicti.keys(), columns=dicti.keys())
for aa_i in dicti:
    for aa_j in dicti:
        int_pot[aa_i][aa_j] = (dicti[aa_i]["lambda"]+dicti[aa_j]["lambda"])/2
```
