```python
import numpy as np
import mdtraj as md
import time
import datetime
import matplotlib.pyplot as plt
```

```python
def find_rg(traj, topo):
    ti = time.time()
    traj = md.load_lammpstrj(traj, top=topo)
    rg = md.compute_rg(traj)
    print(f'Time expended : {datetime.timedelta(seconds=time.time()-ti)}')    
    return traj, rg
```

```python
xtc = '/home/adria/perdiux/test/lammps/compute/A-syn_HPS_100000_frames.atom'
topo = '/home/adria/perdiux/test/lammps/compute/hps_trj.pdb'

traj, rg = find_rg(xtc, topo)
```

```python
plt.figure(num=None, figsize=(8,6))
plt.xlabel("Step")
plt.ylabel("Rg")
plt.title("Rg Convergence for CPEB4\u03944")
plt.loglog(rg, marker=".", linestyle=None);
```

```python
plt.figure(num=None, figsize=(8,6))
plt.xlabel("Step")
plt.ylabel("Rg")
plt.title("Rg Convergence for CPEB4\u03944")
plt.semilogy(rg);
```
