```python
# DATA

wt = analysis.Analysis(oliba_wd='/home/adria/data/prod/lammps//EXT-CPEB4x2')
d7 = analysis.Analysis(oliba_wd='/home/adria/data/prod/lammps/TEST/EXT-7D_CPEB4x2')
wt_means, d7_means = [], []
wt_coms = wt.chain_coms()
d7_coms = d7.chain_coms()
wt_hist = [np.linalg.norm(wt_coms[:, 0, :, :]-wt_coms[:, 1, :, :], axis=1)]
d7_hist = [np.linalg.norm(d7_coms[:, 0, :, :]-d7_coms[:, 1, :, :], axis=1)]


d4 = analysis.Analysis(oliba_wd='/home/adria/data/prod/lammps/TEST/EXT-CPEB4_D4x2')
d12 = analysis.Analysis(oliba_wd='/home/adria/data/prod/lammps/TEST/EXT-12D_CPEB4_D4x2')
d4_means, d12_means = [], []
d4_hist, d12_hist = [], []
d4_coms = d4.chain_coms()
d12_coms = d12.chain_coms()
d4_hist = [np.linalg.norm(d4_coms[:, 0, :, :]-d4_coms[:, 1, :, :], axis=1)]
d12_hist = [np.linalg.norm(d12_coms[:, 0, :, :]-d12_coms[:, 1, :, :], axis=1)]


mbox = 0
for key in d4.box:
    mbox += np.abs(np.array(d4.box[key], dtype='float')).sum()
mbox=mbox/3
cutoff = 100000

for T in range(12):
    d = np.linalg.norm(wt_coms[T, 0, :, :]-wt_coms[T, 1, :, :], axis=1)
    true_d = d[np.where(d < cutoff)].mean()
    wt_means.append(true_d)
    
    d = np.linalg.norm(d7_coms[T, 0, :, :]-d7_coms[T, 1, :, :], axis=1)
    true_d = d[np.where(d < cutoff)].mean()
    d7_means.append(true_d)
    
    d = np.linalg.norm(d4_coms[T, 0, :, :]-d4_coms[T, 1, :, :], axis=1)
    true_d = d[np.where(d < cutoff)].mean()
    d4_means.append(true_d)
    
    d = np.linalg.norm(d12_coms[T, 0, :, :]-d12_coms[T, 1, :, :], axis=1)
    true_d = d[np.where(d < cutoff)].mean()
    d12_means.append(true_d)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.set_xlabel("Temperature (K)")
ax.set_ylabel("<Distance between chains>")
ax.plot(wt.get_temperatures(), wt_means, '.-', label='WT')
ax.plot(d7.get_temperatures(), d7_means, '.-', label="7D")
ax.legend()

fig, ax = plt.subplots(figsize=(8,6))
ax.set_xlabel("Temperature (K)")
ax.set_ylabel("<Distance between chains>")
ax.plot(d4.get_temperatures(), d4_means, '.-',  label='D4')
ax.plot(d12.get_temperatures(), d12_means, '.-', label="12D")
ax.legend()
```
