```python
def mc_opt(T, phi_b_tot, nt_total, repeats=10, steps=1000):
        
    def c_unif(max_val):
        a = random.uniform(0,max_val)
        b = max_val - a
        return np.array([a,b])   
      
    def mc_move(nk_a_i_old, nk_b_i_old, G_min, T, P):

        kick1 = random.uniform(0, np.max(nk_a_i_old))
        kick2 = random.uniform(0, np.max(nk_b_i_old))
        
        if nk_a_i_old[1] > nk_a_i_old[0]: kick1 *= -1
        if nk_b_i_old[1] > nk_b_i_old[0]: kick2 *= -1
        
        n_a_i_new = [nk_a_i_old[0]-kick1, nk_a_i_old[1]+kick1]
        n_b_i_new = [nk_b_i_old[0]-kick2, nk_b_i_old[1]+kick2]
        
        nt_i = [n_a_i_new[0]+n_b_i_new[0], n_a_i_new[1]+n_b_i_new[1]]
        
        phi_a_i = [n_a_i_new[0]/nt_i[0], n_a_i_new[1]/nt_i[1]]
        phi_b_i = [n_b_i_new[0]/nt_i[0], n_b_i_new[1]/nt_i[1]]
        
        G_t = 0
        for phase in range(2):
            G_t += G(phi_a_i[phase], phi_b_i[phase], P[phase], nt_i[phase], T)
#         beta = T*cnt.R
        beta = 150
        
        dG = G_t-G_min    
#         print(dG, np.exp(-dG/beta))
        
        if dG < 0:
            G_min = G_t
            nk_a_i_old[:] = n_a_i_new[:]
            nk_b_i_old[:] = n_b_i_new[:]
        elif rand() < np.exp(-dG/beta):
            G_min = G_t
            nk_a_i_old[:] = n_a_i_new[:]
            nk_b_i_old[:] = n_b_i_new[:]
            
        return nk_a_i_old, nk_b_i_old, G_min
        
    phi_a_tot = 1 - phi_b_tot
    
    nk_a_i = c_unif(nt_total*phi_a_tot)
    nk_b_i = c_unif(nt_total*phi_b_tot)        
    P = [0.5, 0.5]

    G_data = []
    nk_data = []
    finals = []
    for guess in range(repeats):
        nk_a_i = c_unif(nt_total*phi_a_tot)
        nk_b_i = c_unif(nt_total*phi_b_tot)   
        nk_sa, nk_sb = None, None
        G_min = 1e9   
        G_old=0
        for st in range(steps):
            nk_a_i, nk_b_i, G_min = mc_move(nk_a_i, nk_b_i, G_min, T, P)
            if G_min < G_old:
                G_old = G_min
                nk_sa = nk_a_i
                nk_sb = nk_b_i
        G_data.append(G_old)
        nk_data.append([nk_sa, nk_sb])
        finals.append([nk_sb[0]/(phi_b_tot*nt_total), nk_sb[1]/(phi_b_tot*nt_total)])
    return np.array(G_data), np.array(nk_data),np.array(finals)
```

```python
M2 = 156
f = 8

# Flory parameters
xi_1p  = 0.152
xi_1u = 3.333
xi_pu = 1.868
xi_uu = 3.024


d = mc_opt(T=500, phi_b_tot=0.3, nt_total=1, steps=10000, repeats=200)
print("G", d[0][np.argwhere(d[0]==d[0].min())])
print("n_ks",d[1][np.argwhere(d[0]==d[0].min())])
print("phis",d[2][np.argwhere(d[0]==d[0].min())])
```
