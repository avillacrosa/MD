```python
import numpy as np
import matplotlib.pyplot as plt

from style_vars import *
plt.style.use('default')
plt.style.use('style.mplstyle')
```

```python
#WT-D4
names = ['eps80', 'eps40', 'eps20']
titles = ["$_r=80$", "$_r=40$", "$_r=20$"]

fig, axs = plt.subplots(1,len(names), figsize=(16,6))
for i in range(len(names)):
    wt = np.genfromtxt(f'/home/adria/perdiux/prod/lammps/final/RPA/fg_ucst/bi_wt_{names[i]}.txt')
    d4 = np.genfromtxt(f'/home/adria/perdiux/prod/lammps/final/RPA/fg_ucst/bi_d4_{names[i]}.txt')
    axs[i].plot(wt[:,0], 1/wt[:,1], color=C_W, label='WT')
    axs[i].plot(d4[:,0], 1/d4[:,1], color=C_4, label='$\Delta$4')
    axs[i].legend()
    axs[i].set_title(titles[i])
    axs[i].set_xlabel("$\phi_m$")
axs[0].set_ylabel("Interaction strength")
fig.tight_layout()
```
