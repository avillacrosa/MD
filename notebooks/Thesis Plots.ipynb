```python
%load_ext autoreload
%autoreload 2

import scipy.constants as cnt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lattice_F(phi, Na, Nb, T, xi):
    f = phi*np.log(phi)/Na+(1-phi)*np.log(1-phi)/Nb+xi*phi*(1-phi)
    f *= T
    return f
```

```python
temps = np.linspace(0,500,8)
# temps = [250,300,350]

phis = np.linspace(0.0001,0.99999,100)
for temp in temps:
    plt.plot(phis, lattice_F(phis,Na=200,Nb=1,T=temp,xi=-100/temp), label=f'{temp:.0f}')
plt.legend()
```
