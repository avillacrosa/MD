```python
deltas = []
scds = []
test = lmp.LMP(oliba_wd='.')
for seq in svs:
    deltas.append((test.pappu_delta(seq, g=5)+test.pappu_delta(seq, g=6))/2)
    scds.append(test.chan_scd(seq))
deltas = np.array(deltas)
deltas = deltas/np.max(deltas)
```

```python
from sklearn.linear_model import LinearRegression


plt.figure(figsize=(8,6))
plt.ylabel("\u03BA")
plt.xlabel("SCD")
plt.ylim(0,1)
plt.xlim(-30,0)
plt.tick_params(right=True, top=True, direction='in')
reg = LinearRegression().fit(np.array(scds).reshape(-1, 1), deltas)
plt.plot(scds,deltas, '.')
plt.plot(scds,np.array(scds)*reg.coef_[0]+reg.intercept_)
```
