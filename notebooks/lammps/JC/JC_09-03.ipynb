```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

```python
def LJ(x):
    U = 4*eps*((sigma/x)**12-(sigma/x)**6)
    return U

def WCA(x):
    U = np.zeros_like(x)
    U[x < (2**(1/6)*sigma)] = LJ(x[x < (2**(1/6)*sigma)])+eps
    return U

def FENE(x):
    U = -K*R0**2*np.log(1-(x/R0)**2)/2
    return U

def HARMONIC(x):
    U = K * (x/R0)**2
    return U
```

```python
plt.figure(figsize=(8, 6))
sigma = 1
eps = 1
r = np.linspace(0.9, 2.5, 1000)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Long Range potential", fontsize=16)
plt.xlabel("$r_{ij}$", fontsize=24)
plt.ylabel("$U(r_{ij}$)", fontsize=24)
T = mlines.Line2D([], [], color='tab:red', marker='o', linestyle='None',
                          markersize=10, label='LJ (T-T)', markeredgecolor='black')
H = mlines.Line2D([], [], color='tab:blue', marker='o', linestyle='None',
                          markersize=10, label='WCA (T-H, H-H)', markeredgecolor='black')
plt.plot(r, LJ(r), linewidth=4, color='tab:red');
plt.scatter(r, LJ(r), s=4, color='tab:red');
plt.plot(r, WCA(r), linewidth=4, color='tab:blue');
plt.axhline(0, ls='--', color='grey', alpha=0.9);
plt.legend(handles=[T,H], fontsize=14);
```

```python
plt.figure(figsize=(8, 6))

K = 30*eps/sigma**2
R0 = 1.5*sigma
rharm = np.linspace(0.75, R0-0.01, 1000)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("r", fontsize=24)
plt.title("Bond Potential", fontsize=16)
plt.ylabel("U(r)", fontsize=24)
plt.plot(rharm, FENE(rharm) + WCA(rharm), label="FENE + WCA", color='purple', linewidth=4);
plt.axhline(0, ls='--', color='grey', alpha=0.9);
plt.legend(fontsize=14);
```
