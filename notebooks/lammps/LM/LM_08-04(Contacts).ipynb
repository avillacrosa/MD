```python
# INTRA CONTACTS (WT vs D7)

fig, ax = plt.subplots(figsize=(16,12))
img = ax.imshow(wt_contacts[0][0]-d7_contacts[0][0], cmap='seismic', vmin=-0.008, vmax=0.008)
cb = fig.colorbar(img, orientation='horizontal', fraction=0.046, pad=0.08)

for p in d7_phosphosites:
    ax.axvline(p, alpha=0.5, color='tab:orange')
ax.invert_yaxis()
```
