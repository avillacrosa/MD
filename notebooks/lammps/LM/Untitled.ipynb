```python
#L0
rgs_l0 = []

for T in range(12):
    tr = md.load(f'/home/adria/data/prod/OLD/CPEB4/L0/0ls-10I-80e/reorder-{T}.xtc', top='/home/adria/data/prod/OLD/CPEB4/L0/0ls-10I-80e/data_trj.pdb')
    rgs_l0.append(md.compute_rg(tr))
```

```python
#L0
rgs_l1 = []

for T in range(12):
    tr = md.load(f'/home/adria/data/prod/OLD/CPEB4/EPS/80/1.0ls-10I-80e/reorder-{T}.xtc', top='/home/adria/data/prod/OLD/CPEB4/EPS/80/1.0ls-10I-80e/data_trj.pdb')
    rgs_l1.append(md.compute_rg(tr))
```

```python
plt.figure(figsize=(8,6))
rgs_l0 = np.array(rgs_l0)
rgs_l1 = np.array(rgs_l1)
temps = [150.0, 170.0, 192.5, 217.5, 247.5, 280.0, 320.0, 362.5, 410.0, 467.5, 530.0, 600.0]
import matplotlib.pyplot as plt
plt.ylabel("Rg (Å)", fontsize=14)
plt.xlabel("T", fontsize=14)
plt.plot(temps, rgs_l0.mean(axis=1)*10,  label="$\lambda_{ij}=0$", color='tab:green')
plt.plot(temps, rgs_l1.mean(axis=1)*10, label="Normal HPS", color='tab:purple')
plt.legend(fontsize=14)
```
