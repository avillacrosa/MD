```python
import mdtraj as md
import os
import numpy as np
import matplotlib.pyplot as plt
import analysis
```

```python
base_dir = '/home/adria/xoriguer/adria/prod/dipole'
os.path.exists(base_dir)
```

```python
temperatures = [300,336,372,409,463,500]
rgs = []
for T in temperatures:
    traj = md.load(os.path.join(base_dir,f'trajectory_{T}.dcd'), top='/home/adria/xoriguer/adria/prod/topo.pdb')
    rgs.append(md.compute_rg(traj)*10.)
rgs = np.array(rgs)

lammps = analysis.Analysis(oliba_wd='/home/adria/perdiux/prod/lammps/dignon/CPEB4/x1/1.0ls-100I-80e')
lrg = lammps.rg().mean(axis=1)

plt.plot(lammps.temperatures, lrg)
plt.plot(temperatures, rgs.mean(axis=1))
```
