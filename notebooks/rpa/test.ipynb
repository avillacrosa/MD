```python
phis = np.linspace(0.001,0.07,40)
dff = []
ff = []
for phi in phis:
    dff.append(fgRPA.fgRPA(phi_m=phi).df())
    ff.append(fgRPA.fgRPA(phi_m=phi).f())    
```

```python

dphi = (phis[1]-phis[0]) 
dffff = (ff-np.roll(ff,1))/dphi
plt.plot(phis[1:], dffff[1:])
plt.plot(phis,dff)

plt.figure()
plt.plot(phis, ff)
```

```python
def s(x):
    return x*np.log(x)

def ds(x):
    return 1 + np.log(x)

x = np.linspace(0.001,0.07,40)

plt.plot(x, s(x))
plt.figure()
dx = x[1] - x[0]
dns = (s(x) - np.roll(s(x),+1))/dx
plt.plot(x[:-1], dns[1:])
plt.plot(x, ds(x))
```
