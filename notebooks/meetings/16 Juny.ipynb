```python
eps = [20,40,50,70]

rgsD4 = []
rgsD4_err = []
rgs12D = []
rgs12D_err = []

for ep in eps:
    rg_calc_D4 = analysis.Analysis(oliba_wd=f'/home/adria/perdiux/prod/lammps/dignon/HPS-T/EpsExp/D4/{ep}').rg()
    rg_calc_12D = analysis.Analysis(oliba_wd=f'/home/adria/perdiux/prod/lammps/dignon/HPS-T/EpsExp/12D/{ep}').rg()
    rgsD4_err.append(analysis.Analysis(oliba_wd=None).block_error(rg_calc_D4))
    rgs12D_err.append(analysis.Analysis(oliba_wd=None).block_error(rg_calc_12D))
    rgsD4.append(rg_calc_D4.mean(axis=1))
    rgs12D.append(rg_calc_12D.mean(axis=1))

# Eps 80
rg_calc_D4 = analysis.Analysis(oliba_wd=f'/home/adria/perdiux/prod/lammps/dignon/HPS-T/0.75/SingleChain/D4').rg()[[2,5,8,11],:]
rg_calc_12D = analysis.Analysis(oliba_wd=f'/home/adria/perdiux/prod/lammps/dignon/HPS-T/0.75/SingleChain/12D').rg()
rgsD4_err.append(analysis.Analysis(oliba_wd=None).block_error(rg_calc_D4))
rgs12D_err.append(analysis.Analysis(oliba_wd=None).block_error(rg_calc_12D))
rgsD4.append(rg_calc_D4.mean(axis=1))
rgs12D.append(rg_calc_12D.mean(axis=1))
```

```python
fig, ax = plt.subplots(figsize=(10,10))
temperatures = np.linspace(280,320,4)
ax.set_title("$\lambda$=0.75")
eps = [20,40,50,70, 80]
for i, ep in enumerate(eps):
    p = ax.errorbar(temperatures, rgsD4[i], yerr=rgsD4_err[i], label=f"ε = {ep}", capsize=3)
    ax.errorbar(temperatures, rgs12D[i], yerr=rgs12D_err[i], color=p[0].get_color(), ls='--', capsize=3)
ax.axhline(440**0.5*5.5/6**0.5, color='black', linestyle='-.')
legend_elements = [Line2D([0], [0], ls='--', label='12D', color='black'),
                   Line2D([0], [0], ls='-', label='D4', color='black')]
l2 = ax.legend(handles=legend_elements, loc='upper left', fontsize=14)
ax.add_artist(l2)
ax.set_xlabel("T(K)", fontsize=14)
ax.set_ylabel("Rg ($\AA$)", fontsize=14)
ax.legend(fontsize=14)
```
