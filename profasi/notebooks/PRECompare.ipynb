# PRE comparison

Import statements

```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

Load all PRE's : experimental, PROFASI's and LAMMPS

```python
#Paths relative to oliba

profasi_path = '/home/adria/data/pres/asyn/calc_pre'
exp_path     = '/home/adria/data/pres/asyn/exp_pre'
lammps_path  = '/home/adria/data/pres/asyn/calpha_pre'

# labels = [20, 85]
labels = [18, 20, 24, 42, 62, 76, 85]
labels = [18, 20, 24, 42, 62, 76, 85, 87, 90, 103, 120, 140]


profasiPREs  = []
lammpsPREs   = []
expPREs      = []

for lab in labels:
    plt.figure()
    
    prf = np.genfromtxt(os.path.join(profasi_path,"1_100000-res-{}.dat".format(lab)))
    exp = np.genfromtxt(os.path.join(exp_path,"exp{}.dat".format(lab)))
    lammps = np.genfromtxt(os.path.join(lammps_path,"lammps-{}.dat".format(lab)))
    
    profasiPREs.append(prf)
    expPREs.append(exp)
    lammpsPREs.append(lammps)
    
    plt.title(f"Label {lab:d}")
    plt.plot(prf[:,0], prf[:,1], label="PROFASI")
    plt.plot(lammps[:,0], lammps[:,1], label="C\u03B1 Model (LAMMPS)")
    plt.plot(prf[:,0], exp, label="Experimental")

    plt.legend()
```
